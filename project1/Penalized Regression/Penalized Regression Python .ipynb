# Penalised Regression in Python

### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
```

### Loading and Splitting the data

```python
# Loading data
data = pd.read_csv('QuestionMark.csv')
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
w4 = 'w4'
encoder = LabelEncoder()
encoder.fit(data[w4])
data[w4] = encoder.transform(data[w4])
X = data.drop(['y'], axis=1)
y = data['y']

# Spliting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Ridge Regression

```python
ridge_model = Ridge(alpha = 0.0)
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
```

### Calculating the RMSE

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE = ",rmse)
```

### Calculating Coefficient of Determination 

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
print("Coefficient of Determination = ", r2)
```

```python
from sklearn.linear_model import Lasso
```

# LASSO 

```python
lasso_reg = Lasso(alpha=1.0) 
lasso_reg.fit(X_train, y_train)
y_pred = lasso_reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('RMSE = ', rmse)
r2 = r2_score(y_test, y_pred)
print("Coefficient of Determination = ", r2)
```

## ELASTIC NET

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
elastic_model = ElasticNet()
param_grid = {'alpha' : np.linspace(-4,-4,10),
             'l1_ratio' : np.linspace(0,1,10)}
grid_search = GridSearchCV(elastic_model, param_grid, cv = 10)
grid_search.fit(X_train, y_train)
y_pred = grid_search.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('RMSE = ', rmse)
r2 = r2_score(y_test, y_pred)
print("Coefficient of Determination = ", r2)
print("Best hyperparamters : ", grid_search.best_params_)
```
